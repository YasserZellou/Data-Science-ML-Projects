### Process the 'Fare' feature

1) Check if there are any missing values in `Fare`

2) If there are missing values, fill them with the _median_ value of `Fare`

```python
import pandas as pd


data = pd.read_csv('titanic.csv')
col = 'Fare'  

missing = data[col].isnull().sum()

if missing > 0:
    
    print(f"'{col}' is missing {missing} values")
    
    med = data[col].median()
    
    data[col].fillna(med, inplace=True)
    
    print(f"'{col}' missing values have been replaced with {med}.")
```

### Task 8
Draw a histogram of survival rates as a function of `Pclass`.

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('titanic.csv')

surviv = data.groupby('Pclass')['Survived'].mean()
plt.bar(surviv.index, surviv.values)
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.title('Survival Rate')
plt.show()
```
